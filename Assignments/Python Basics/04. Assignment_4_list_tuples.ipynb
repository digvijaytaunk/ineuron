##### 17. How do you distinguish between copy.copy() and copy.deepcopy()?

#### Ans. 
For a nested list or a complex list `copy.copy()` functon make a new object and insert the reference elements in the original list. Hence if the elemets of original list are updated the shallow copied object will also be updated. 
Whereas in `copy.deepcopy()`, a new object is created with a seperate copy of all the nested elements inserted into it with no relation/connection wiht the original list object.
